```
df_gold = gold[['rates','date']]
df_gold = df_gold.rename(columns = {"date":"ds","rates":"y"})
df_gold.head(5)
```

```
m = Prophet(daily_seasonality = True) # the Prophet class (model)
m.fit(df_gold) # fit the model using all data
```

```
future_price = m.make_future_dataframe(periods=60) #we need to specify the number of days in future
prediction = m.predict(future_price)
prediction[['ds','yhat','yhat_lower','yhat_upper']].head()
fig = m.plot(prediction)
#plt.title("Prediction of the Gold Price using the FB Prophet")
#plt.xlabel("Date")
#plt.ylabel("Value")
#plt.show()
```

```
from prophet.plot import add_changepoints_to_plot
a = add_changepoints_to_plot(fig.gca(),m,prediction)
fig
```

```
#### Improving the Model
m_2 = Prophet(n_changepoints=20, monthly_seasonality=True)
m_2.fit(df_gold) # fit the model using all data
```

```
future_price = m_2.make_future_dataframe(periods=120) #we need to specify the number of days in future
prediction = m_2.predict(future_price)
prediction[['ds','yhat','yhat_lower','yhat_upper']].head()
fig = m_2.plot(prediction)
#plt.title("Prediction of the Gold Price using the FB Prophet")
#plt.xlabel("Date")
#plt.ylabel("Value")
#plt.show()
```
